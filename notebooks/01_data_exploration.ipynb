### Task 2: The Data Loader
**File:** `src/data_loader.py`

Move away from writing raw `pd.read_csv` in notebooks. We need a class to handle this.

### Task 3: The Model Interface (Abstract Base Class)
**File:** `src/models/base.py`

In MLE, we need standard interfaces so we can swap models easily (e.g., switching from Popularity to Deep Learning without breaking the API).

1.  **Define the Class**:
    * Create a class `BaseRecommender` that inherits from `ABC` (Abstract Base Class).
2.  **Define Abstract Methods**:
    * `fit(self, train_data)`: Should perform model training.
    * `recommend(self, user_id, n)`: Should return a list of top-N items.
    * `evaluate(self, test_data)`: (Optional) For calculating metrics.

> **Action**: Open `src/models/base.py` and implement the abstract base class using the `abc` module.

---

### Task 4: The Baseline Model (Popularity)
**File:** `src/models/popularity.py`

Implement a simple "Most Popular" logic. This will serve as our baseline to compare against future complex models.

1.  **Inheritance**:
    * Create class `PopularityRecommender` inheriting from `BaseRecommender`.
2.  **Implement `fit`**:
    * Logic: Group the training data by `movie_id`, count the number of ratings, and sort in descending order.
    * Store the top items in a class attribute (e.g., `self.popular_items`).
3.  **Implement `recommend`**:
    * Logic: Simply return the top-N items stored in `self.popular_items`.
    * *Challenge*: Since this is non-personalized, the `user_id` input is technically not used for calculation, but keep it in the signature to match the interface.

> **Action**: Open `src/models/popularity.py` and implement the logic.
